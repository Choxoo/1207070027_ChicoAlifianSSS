```python
# Untuk meng-import library yang tersedia di python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
# untuk mendeklarasikan gambar
img = imageio.imread("1.jpg")
```

```python
img_height = img.shape[0] # Mengambil tinggi dari gambar
img_width = img.shape[1] # Mengambil lebar dari gambar
img_channel = img.shape[2] # Mengambil warna dari gambar
img_type = img.dtype # Mengambil type gambar
```

# Brightness Grayscale


```python
img_brightness = np.zeros(img.shape, dtype=np.uint8) # untuk meng kosongkan array yg sudah ada pada gambar
```

```python
def brighter(nilai): # fungsi brighter dengan parameter nilai
    for y in range(0, img_height): # Looping untuk tinggi gambar 
        for x in range(0, img_width): # Looping untuk lebar gambar
            red = img[y][x][0]
            green = img[y][x][1] # untuk mendapatkan informasi piksel gambar dari RGB
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3 # menghitung nilai rata-rata untuk mendapatkan nilai keabuan
            gray += nilai
            if gray > 255: # nilai batas
                gray = 255
            if gray < 0: # nilai batas
                gray = 0
            img_brightness[y][x] = (gray, gray, gray) # untuk menampung nilai baru
```

```python
brighter(-100) # memanggil fungsi brighter dengan nilai -100
plt.imshow(img_brightness) # menampilkan gambar
plt.title("Brightness -100") # menampilkan title gambar
plt.show() # menampilkan plot

brighter(100) # memanggil fungsi brighter dengan nilai 100
plt.imshow(img_brightness) # menampilkan gambar
plt.title("Brightness 100") # menampilkan title gambar
plt.show()# menampilkan plot
```

# Brightness RGB


```python
img_rgbbrightness = np.zeros(img.shape, dtype=np.uint8) # untuk meng kosongkan array yg sudah ada pada gambar
```

```python
def rgbbrighter(nilai): # fungsi rgbbrighter dengan parameter nilai
    for y in range(0, img_height): # Looping untuk tinggi gambar 
        for x in range(0, img_width): # Looping untuk Lebar gambar 
            red = img[y][x][0] # untuk mendapatkan informasi piksel gambar dari red
            red += nilai
            if red > 255: # nilai batas
                red = 255
            if red < 0: # nilai batas
                red = 0
            green = img[y][x][1] # untuk mendapatkan informasi piksel gambar dari green
            green += nilai
            if green > 255: # nilai batas
                green = 255
            if green < 0: # nilai batas
                green = 0
            blue = img[y][x][2] # untuk mendapatkan informasi piksel gambar dari blue
            blue += nilai
            if blue > 255: # nilai batas
                blue = 255
            if blue < 0: # nilai batas
                blue = 0
            img_rgbbrightness[y][x] = (red, green, blue) #  untuk menampung nilai baru
```

```python
rgbbrighter(-100) # memanggil fungsi rgbbrighter dengan nilai -100
plt.imshow(img_rgbbrightness) # menampilkan gambar
plt.title("Brightness -100") # menampilkan title gambar
plt.show() # menampilkan plot

rgbbrighter(100) # memanggil fungsi rgbbrighter dengan nilai 100
plt.imshow(img_rgbbrightness) # menampilkan gambar
plt.title("Brightness 100") # menampilkan title gambar
plt.show() # menampilkan plot
```

# Contrass


```python
img_contrass = np.zeros(img.shape, dtype=np.uint8)# untuk meng kosongkan array yg sudah ada pada gambar
```

```python
def contrass(nilai): # fungsi contrass dengan parameter nilai
    for y in range(0, img_height): # Looping tinggi gambar
        for x in range(0, img_width): # Looping lebar gambar
            red = img[y][x][0]
            green = img[y][x][1] # untuk mendapatkan informasi piksel gambar dari RGB
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3 # menghitung nilai rata-rata untuk mendapatkan nilai keabuan
            gray *= nilai # mengalikan nilai keabuan dengan nilai
            if gray > 255: # nilai batas
                gray = 255
            img_contrass[y][x] = (gray, gray, gray) #  untuk menampung nilai baru
```

```python
contrass(2)# memanggil fungsi contrass dengan nilai 2
plt.imshow(img_contrass) # meanmpilkan gambar
plt.title("Contrass 2") # menampilkan title
plt.show() # menampilkan plot

contrass(3) # memanggil fungsi contrass dengan nilai 3
plt.imshow(img_contrass) # menampilkan gambar
plt.title("Contrass 3") # menampilkan title
plt.show() # menampilkan plot
```

# Contrass Auto Level

```python
img_autocontrass = np.zeros(img.shape, dtype=np.uint8) # untuk meng kosongkan array yg sudah ada pada gambar
```

```python
def autocontrass(): # fungsi contrasss
    xmax = 300 # mendefinisikan variabel xmax dengan 300
    xmin = 0 # mendefinisikan variabel xmin dengan 0
    d = 0 # mendefinisikan variabel d dengan 0
    # Mendapatkan nilai d, dimana nilai d ini akan berpengaruh pada hitungan
    # untuk mendapatkan tingkat kontras
    for y in range(0, img_height): # Looping tinggi gambar 
        for x in range(0, img_width): # Looping lebar gambar
            red = img[y][x][0]
            green = img[y][x][1] # untuk mendapatkan informasi piksel gambar dari RGB
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3 # menhitung nilai rata-rata untuk mendapatkan nilai keabuan
            if gray < xmax: # nilai batas
                xmax = gray
            if gray > xmin: # nilai batas
                xmin = gray
    d = xmin-xmax # menghitung selisih xmin-xmax dan menyimpannya di variable d
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray = int(float(255/d) * (gray-xmax)) # menghitung nilai keabuan baru
            img_autocontrass[y][x] = (gray, gray, gray) #  untuk menampung nilai baru
```

```python
autocontrass() # memanggil fungsi auto contrass
plt.imshow(img_autocontrass) # menampilkan gambar
plt.title("Contrass Autolevel") # menampilkan title
plt.show() # menampilkan plot
```
