```python
# Meng import library yg ada pada python
import matplotlib.pyplot as plt
import cv2

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.util import invert
import numpy as np
```

# Percobaan 1 - Cropping Image

```python
awanImage = imread (fname='1.jpg') # untuk memuat Gambar
kucingImage = imread(fname="2.jpg") # untuk memuat Gambar 

awanCropped = awanImage.copy() # Meng copy lalu meng cropping gambar
awanCropped = awanCropped[64:256,90:1080]


kucingCropped = kucingImage.copy() # Meng copy lalu meng cropping gambar
kucingCropped = kucingCropped[64:256,30:480]

print('Awan Ori Shape : ',awanImage.shape) # menampilkan dimensi asli
print('Awan Crop Shape : ',awanCropped.shape) # menampilkan dimensi yg sudah dicrop

fig, axes = plt.subplots(2, 2, figsize=(12, 12)) # membuat subplot dengan 2 baris dan 2 kolom dengan jarak 12
ax = axes.ravel()

ax[0].imshow(awanImage) # menampilkan gambar ori
ax[0].set_title("Citra Input 1") # menampilkan title

ax[1].imshow(kucingImage, cmap='gray') # menampilkan gambar dengan skala gray
ax[1].set_title('Citra Input 2') # menampilkan title

ax[2].imshow(awanCropped) # menampilkan hasil gambar cropped
ax[2].set_title("Citra Output 1") # menampilkan title

ax[3].imshow(kucingCropped, cmap='gray') # menampilkan hasil gambar cropped dengan skala gray
ax[3].set_title('Citra Output 2') # menampilkan title
```

# Percobaan 2 - Citra Negative

```python
inv = invert(awanCropped) # melakukan invert pada citra awan yg sudah di cropped
print('Shape Input : ', awanCropped.shape) #menampilkan dimensi citra hasil cropped
print('Shape Output : ',inv.shape) #menampilkan dimensi citra hasil invert

fig, axes = plt.subplots(2, 2, figsize=(12, 12)) # membuat subplot dengan 2 baris dan 2 kolom dengan jarak 12
ax = axes.ravel()

ax[0].imshow(awanCropped) # Menampilkan citra awal yg di cropped
ax[0].set_title("Citra Input")

ax[1].hist(awanCropped.ravel(), bins=256) # menampilkan histogram citranya
ax[1].set_title('Histogram Input')

ax[2].imshow(inv) # menampilkan citra hasil invert
ax[2].set_title('Citra Output (Inverted Image)')

ax[3].hist(inv.ravel(), bins=256) # menampilkan histogram citranya
ax[3].set_title('Histogram Output')
```

```python
imggray = rgb2gray(kucingCropped) # mengubah citra menjadi grayscale

print('img2gray shape = ',imggray.shape) # menampilkan dimensi citra

fig, axes = plt.subplots(1, 2, figsize=(12, 12)) # membuat subplot dengan 2 baris dan 2 kolom dengan jarak 12
ax = axes.ravel()

ax[0].imshow(kucingCropped) # menampilkan hasil citra cropped 
ax[0].set_title("img2 RGB Image")
ax[1].imshow(imggray,cmap=plt.cm.gray) # menampilkan hasil citra cropped yg sudah grayscale
ax[1].set_title("img2 Grayscale Image")

fig.tight_layout() # mengatur tata letak subplot
plt.show() # menampilkan subplot
```

```python
copyCamera = imggray.copy().astype(float) # menyalin citra grayscale dan mengubah type datanya

m1,n1 = copyCamera.shape # untuk mendapatkan dimensi citra yang disalin
output1 = np.empty([m1, n1]) # membuat array kosong

for baris in range(0, m1-1): # melakukan pengolahan pada setiap piksel citra
    for kolom in range(0, n1-1):
        a1 = baris
        b1 = kolom
        output1[a1, b1] = copyCamera[baris, kolom] + 10
        
fig, axes = plt.subplots(2, 2, figsize=(12, 12)) # membuat subplot dengan 2 baris dan 2 kolom dengan jarak 12
ax = axes.ravel()

ax[0].imshow(copyCamera, cmap='gray') # menampilkan citra dalam format keabuan
ax[0].set_title("Citra Input")

ax[1].hist(copyCamera.ravel(), bins=256) # menampilkan histogram 
ax[1].set_title('Histogram Input')

ax[2].imshow(output1, cmap='gray') # menampilkan citra hasil penambahan brightness
ax[2].set_title('Citra Output (Brightnes)')

ax[3].hist(output1.ravel(), bins=256) # menampilkan histogramnnya
ax[3].set_title('Histogram Input')
```
